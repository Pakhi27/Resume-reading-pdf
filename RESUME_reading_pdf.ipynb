```python
!pip install reportlab

from reportlab.lib.pagesizes import letter
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.enums import TA_CENTER
from IPython.display import display, FileLink

def clean_and_format_text(extracted_text):
    cleaned_text = extracted_text.replace("", "-").replace("", "-")
    lines = cleaned_text.split('\n')
    formatted_sections = []
    section_headings = ["PROFILE", "EDUCATION", "PROJECTS", "SKILLS", "ACHIEVEMENTS & ACTIVITIES", "AWARDS"]

    for line in lines:
        line = line.strip()
        if line:
            if any(heading in line for heading in section_headings):
                formatted_sections.append(('section_heading', line))
            elif line.startswith("- "):
                formatted_sections.append(('project_name', line))
            else:
                formatted_sections.append(('normal', line))

    return formatted_sections

def create_pdf(formatted_sections, pdf_path):
    doc = SimpleDocTemplate(pdf_path, pagesize=letter)
    styles = getSampleStyleSheet()

    centered_style = ParagraphStyle(name="centered", parent=styles['Normal'], alignment=TA_CENTER, fontSize=14, leading=16)
    section_heading_style = ParagraphStyle(name="section_heading", parent=styles['Normal'], fontSize=12, leading=14, spaceBefore=12, spaceAfter=12)
    project_name_style = ParagraphStyle(name="project_name", parent=styles['Normal'], fontSize=11, leading=13, spaceBefore=8, spaceAfter=8, textColor='blue')
    normal_style = styles['Normal']

    header = [
        Paragraph("<b>Pakhi Singhal</b>", centered_style),
        Paragraph("Phone: 8401973442", normal_style),
        Paragraph("Dharwad, Karnataka", normal_style),
        Spacer(1, 0.2 * inch)
    ]

    paragraphs = []
    for style, text in formatted_sections:
        if style == 'section_heading':
            paragraphs.append(Paragraph(text, section_heading_style))
        elif style == 'project_name':
            paragraphs.append(Paragraph(text, project_name_style))
        else:
            paragraphs.append(Paragraph(text, normal_style))
        paragraphs.append(Spacer(1, 0.2 * inch))

    doc.build(header + paragraphs)

extracted_text = """
Dharwad, Karnataka
PROFILE
I am a B.Tech 3rd year student in Data Science and AI at IIIT Dharwad. Enthusiastic about web development and AL/ML,
aiming to blend technology and creativity for innovative solutions. Passionate about contributing to impactful projects
at the intersection of data-driven insights, web development and Machine Learning.
EDUCATION
B.Tech in Data Science and Artificial Intelligence, IIIT DHARWAD
2022-2026
High School, New era public school-secured 3rd rank in the district
2020 96,4%
Senior Secondary, Nalanda Vidhyalya
PROJECTS
2020-2022 93.4%
- Mall -Customer Dataset-Unsupervised Machine Learning
- Analyzed customer purchase data from a mall dataset using unsupervised machine learning techniques:
- Employed clustering algorithms (K-Means, Hierarchical, DBSCAN) to segment customers based on purchasing habits.
- Applied PCA for dimensionality reduction and anomaly detection to identify unusual purchase patterns.
- Resulted in the identification of two distinct customer clusters.
- YouTube Toxic Comment Classification
- Developed a Neural Network model to classify toxic comments on YouTube:
- Utilized RNN or LSTM networks to handle the sequential nature of text data.
- Achieved 100% accuracy in toxic comment classification, enhancing user experience by filtering harmful content.
- Pre-processed comment data by removing punctuation, stop words, and converting text to numerical features.
- Iris Flower Prediction with Machine Learning
- Analyzed the Iris flower dataset to classify flower species:
- Implemented KNN, Logistic Regression, Decision Tree Classifier, SVM, and Naive Bayes algorithms.
- Achieved perfect accuracy with KNN, Logistic Regression, DTC, and SVM, and 97% accuracy with Naive Bayes.
- California Housing Price Prediction with Machine Learning (Regression)
- Predicted California housing prices using regression techniques:
- Utilized Linear Regression, Decision Tree Regressor, Random Forest Regressor, Ridge, and Lasso Regression.
- Achieved 78.5% accuracy with Random Forest Regression in predicting house prices.
- IIIT-Dharwad-UI Clone
- Recreated IIIT Dharwad's UI using HTML, CSS, and JavaScript, ensuring responsiveness across devices.
- Flipkart-UI Clone
- Revamped Flipkart's UI with modern design trends using HTML, CSS, and JavaScript, focusing on enhanced user
interactions and cross-browser compatibility.
- Tic-Tac-Toe, Calculator, Alarm Clock
- Developed a Tic-Tac-Toe game, Calculator, and Alarm Clock, showcasing proficiency in front-end development,
responsive design, and feature integration.
SKILLS
C/C++ | HTML | CSS | Javascript | GIT | DBMS | SQL | OS | Python | Data Analysis | AI | ML | DL
ACHIEVEMENTS & ACTIVITIES
- Problem-solving skills: Achieved 2-star rating on CodeChef and active participant in coding contests.
- Education: Cleared JEE with 95.5 percentile; ranked 3rd in district in class 10th with 96.4%.
- Extracurricular Activities:
  - Singing: Actively participate in college music events.
  - Dancing: Involved in college dance club performances.
  - Volleyball: Player in college team; participated in inter IIIT volleyball tournament.
- Music and Dance Club: Member contributing to performances and events.
AWARDS
- Received the Academic Excellence Award for outstanding academic performance in 12th Standard and earned
recognition in the Inter-House Volleyball Tournament for exceptional teamwork and sportsmanship.
"""

formatted_sections = clean_and_format_text(extracted_text)
pdf_path = "/content/Formatted_CV.pdf"
create_pdf(formatted_sections, pdf_path)

display(FileLink(pdf_path))
print("PDF created successfully. Check the file at", pdf_path)


```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

pdf_path = '/content/Pakhi_Singhal_CV(Final).pdf'
text = extract_text_from_pdf(pdf_path)

```

```python
import pandas as pd
from io import StringIO

def parse_text_to_csv(text):
    # Split text into lines
    lines = text.split('\n')

    # Create a DataFrame from lines (assumes each line is a row in CSV)
    df = pd.DataFrame(lines, columns=["Text"])

    return df

df = parse_text_to_csv(text)
```

```python
csv_path = '/content/extracted_data.csv'
df.to_csv(csv_path, index=False)
```
